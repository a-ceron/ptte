Realiza una función que reciba como entrada dos de los elementos jurídicos (elementoJuridicoInicial, elementoJuridicoFinal y como salida de una lisa de elementos que muestre la ruta de conexión de los dos elementos dados).

```python
input(elemento1, elemento5) -> output: [elemento1, elemento3, elemento5]
```

## Fuente de datos. 

Lista de tuplas dadas.

## Resumen de la implementación.

Se generará un grafo que conecte a todas las tuplas, mediante sus relaciones, después se realizara una búsqueda para determinar los elementos que le relacionan. 

````python
import networkx as nx
import matplotlib.pyplot as plt
````

````python
def crear_grafo(relaciones):
    grafo = nx.Graph()
    grafo.add_edges_from(relaciones)
    return grafo

def buscar_camino(grafo, nodo_inicial, nodo_final):
    if not grafo.has_node(nodo_inicial) or not grafo.has_node(nodo_final):
        return []

    cola = [(nodo_inicial, [nodo_inicial])]
    
    while cola:
        nodo, camino = cola.pop(0)
        
        if nodo == nodo_final:
            return camino
        
        for vecino in grafo.neighbors(nodo):
            if vecino not in camino:
                cola.append((vecino, camino + [vecino]))
    
    return []
````

````python
# Por implementar: 
# Una verificación para que se acepte cualquier
# entrada de texto, permitiendo errores gramaticales.
relaciones = [
    ('Principio de Contradiccion', 'Sentencia 338/2018'),
    ('Juez de Amparo', 'Tribunal Colegiado'),
    ('Auto de Vinculacion a Proceso', 'Juez de Control'),
    ('Organo Jurisdiccional', 'Juez de Control'),
    ('Juzgado de Control', 'Sentencia 338/2018'),
    ('Juzgado de Distrito', 'Sentencia 338/2018'),
    ('Norma Convencional que contempla justa indemnizacion','Sentencia 423/2019'),
    ('Autoridad Jurisdiccional', 'Suprema Corte de Justicia de la Nacion'),
    ('Organo Jurisdiccional', 'Juez de Amparo'),
    ('Norma Convencional que contempla justa indemnizacion','Articulo 63.1 Convencion Americana sobre Derechos Humanos'),
    ('Juez de Amparo', 'Juzgado de Distrito'),
    ('Tribunal Colegiado', 'Sentencia 338/2018'),
    ('Norma Convencional que contempla justa indemnizacion','Norma Convencional'),
    ('Prision Preventiva Oficiosa', 'Organo Jurisdiccional'),
    ('Autoridad Jurisdiccional', 'Organo Jurisdiccional')
]
````

````python
# Crear el grafo a partir de las relaciones
grafo = crear_grafo(relaciones)

# Dibujar el grafo
plt.figure(figsize=(20,10))
nx.draw(grafo, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
````

````python
# Calculamos el camino, note que para ciertas variables no 
# existe un camino.
camino = buscar_camino(grafo, 'Auto de Vinculacion a Proceso', 'Suprema Corte de Justicia de la Nacion')
print(camino)
````
